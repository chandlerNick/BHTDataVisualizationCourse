```python
#remove non independent countries
#set 'Country' as the index
non_countries = [
     'Akrotiri','American Samoa' , 'Antarctica', 'Antigua and Barbuda','Ashmore and Cartier Islands',
     'Baker Island', 'Bouvet Island', 'British Indian Ocean Territory', 'British Virgin Islands', 
     'Cayman Islands', 'Christmas Island', 'Clipperton Island', 'Cocos (Keeling) Islands', 'Coral Sea Islands',
     'Dhekelia', 'Falkland Islands (Islas Malvinas)', 'Faroe Islands', 'French Polynesia', 'French Southern and Antarctic Lands',
     'Gibraltar', 'Guernsey' , 'Heard Island and McDonald Islands', 'Howland Island','Isle of Man','Jan Mayen',
     'Jarvis Island', 'Jersey', 'Johnston Atoll','Kingman Reef','Macau' , 'Midway Islands', 'Montserrat','Navassa Island',
     'New Caledonia','Niue', 'Norfolk Island','Northern Mariana Islands', 'Palmyra Atoll', 'Paracel Islands','Pitcairn Islands',
     'Saint Barthelemy', 'Saint Helena, Ascension, and Tristan da Cunha', 'Saint Martin', 'Saint Pierre and Miquelon','Sint Maarten',
     'Spratly Islands', 'Svalbard', 'Tokelau', 'Turks and Caicos Islands', 'Virgin Islands' ,'Wake Island', 'Wallis and Futuna'
]

wfb_df.set_index('Country', inplace=True)
wfb_df.drop(non_countries, axis=0, inplace=True)
print(wfb_df.info())
wfb_df.head()
```

#### Change column 'Geography: Geographic coordinates' to Float columns('Geography: Latitude', 'Geography: Longitude')

```python
#make copy for string manipulation
location_df = wfb_df[['Geography: Geographic coordinates']]
#remove comma deperating latitude and longitude coordinates
location_df['Geography: Geographic coordinates'] = location_df['Geography: Geographic coordinates'].str.replace(',','')
#get degree, minutes, and direction
location_df[['lat_degree', 'lat_minutes', 'lat_direction','long_degree', 'long_minutes', 'long_direction']] = location_df['Geography: Geographic coordinates'].str.split(' ',expand=True)

#calculate latitude decimal coordinates
location_df['lat_decimal']= location_df['lat_degree'].astype(float) + location_df['lat_minutes'].astype(float)/60
location_df.loc[location_df['lat_direction'] == 'S', 'lat_decimal'] *= -1
#save to original data fram
wfb_df['Geography: Latitude'] = location_df['lat_decimal']

#repeat calculation for longitude
location_df['long_decimal']= location_df['long_degree'].astype(float) + location_df['long_minutes'].astype(float)/60
negative_indices = location_df[location_df['long_direction'] == 'S']
location_df.loc[location_df['long_direction'] == 'W', 'long_decimal'] *= -1
wfb_df['Geography: Longitude'] = location_df['long_decimal']

wfb_df[['Geography: Latitude', 'Geography: Longitude']]
```
